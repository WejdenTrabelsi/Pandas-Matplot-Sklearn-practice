```python
from sklearn.cluster import KMeans
#une liste vide pour enregistrer les inerties
intertia_list=[]
#notre liste de nombres de clusters
k_list=range(1,10)
for k in k_list:
    kmeans = KMeans(n_clusters=k) #on dirait tu a crée un objet du KMeans puis il s'entraine avec fit() selon data x ALSO !! n_clusters est un parametre de KMeans 
    #on entraine
    kmeans.fit(x)
    intertia_list.append(kmeans.inertia_)  #variable intertia_ definie dans kmeans dans laquelle on enregistre les inertias
```

```python
plt.plot(k_list,intertia_list)
plt.title("La méthode Elbow")
plt.xlabel("nombre de cluster")
plt.ylabel("Inertie intra-classe")
plt.show()
```
